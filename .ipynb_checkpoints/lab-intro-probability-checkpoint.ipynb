**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
import numpy as np
from scipy.stats import binom

# Given values
total_seats = 450
tickets_sold = 460
prob_no_show = 0.03
prob_show = 1 - prob_no_show

# Random variable:
# Number of passengers who SHOW UP ~ Binomial(n=460, p=0.97)

# We want the probability that at most 450 passengers show up
probability_all_seated = binom.cdf(total_seats, tickets_sold, prob_show)

probability_all_seated
```

**Challenge 2**

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python
from scipy.stats import geom

# Given probability of success on one attempt
p_success = 0.3

# We want: at least 3 attempts before success
# That means success happens on attempt 3 or later
# P(X >= 3) = 1 - P(X <= 2)

prob_at_least_3_attempts = 1 - geom.cdf(2, p_success)

prob_at_least_3_attempts
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
from scipy.stats import poisson

# Given values
avg_visits_per_hour = 500   # λ
server_capacity = 550

# Probability that visits exceed server capacity
# P(X > 550) = 1 - P(X ≤ 550)
prob_server_overwhelmed = 1 - poisson.cdf(server_capacity, avg_visits_per_hour)

prob_server_overwhelmed
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
from scipy.stats import poisson

# Given values
avg_visits_per_hour = 500   # λ
server_capacity = 550
hours_per_day = 24

# Probability server is overwhelmed in one hour
p_overwhelmed_hour = 1 - poisson.cdf(server_capacity, avg_visits_per_hour)

# Probability server is overwhelmed at least once in 24 hours
p_overwhelmed_day = 1 - (1 - p_overwhelmed_hour) ** hours_per_day

p_overwhelmed_day
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
from scipy.stats import expon

# Given arrival rate (customers per minute)
lambda_rate = 1 / 10  # 0.1

# Time window (minutes)
time = 5

# Probability that next arrival is within 5 minutes
prob_within_5_minutes = expon.cdf(time, scale=1/lambda_rate)

prob_within_5_minutes
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
from scipy.stats import expon

# Arrival rate (customers per minute)
lambda_rate = 1 / 10  # 0.1

# Time without arrivals (minutes)
time = 15

# Probability no customer arrives for 15 minutes
prob_employee_break = 1 - expon.cdf(time, scale=1/lambda_rate)

prob_employee_break
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
from scipy.stats import norm

# Given parameters
mean_weight = 150
std_weight = 10

# Probability that weight is between 140 and 160 grams
prob_between_140_160 = norm.cdf(160, mean_weight, std_weight) - norm.cdf(140, mean_weight, std_weight)

prob_between_140_160
```

**Challenge 6**

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
from scipy.stats import expon

# Given mean lifetime (hours)
mean_lifetime = 50
lambda_rate = 1 / mean_lifetime

# Time window (hours)
time = 30

# Probability that the component fails within the first 30 hours
prob_fail_within_30 = expon.cdf(time, scale=1/lambda_rate)

prob_fail_within_30
```
